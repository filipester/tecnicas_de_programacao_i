1) Utilizando o DataFrame importado anteriormente (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd
import numpy as np
```

```python
alunos_3 = pd.read_csv('../dados/alunos3.csv', delimiter=';')
```

```python
alunos_3.head()
```

```python
alunos_3.info()
```

```python
alunos_3[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']] = alunos_3[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].replace(',','.', regex=True).astype(float)
```

```python
# 1
alunos_3['Media_provas'] = alunos_3[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
alunos_3
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
print(alunos_3.loc[alunos_3['Media_provas'].idxmax()])
print('-'*20)
print(alunos_3.loc[alunos_3['Media_provas'].idxmin()])
```

```python
completo.groupby('Sexo')['Media_provas'].mean()
```
