7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
clientes['Total'] = clientes[['Ativo01','Ativo02','Ativo03','Ativo04','Ativo05','Ativo06','Ativo07']].sum(axis=1)
```

```python
clientes['Media'] = clientes[['Nota1', 'Nota2']].mean(axis=1, skipna=True) # Se havia nota NaN, foi ignorada e mantida apenas a nota existente
```

```python
clientes
```

```python
clientes['Dominio'] = clientes['Email'].str.split('@',expand=True)[1]
```

```python
clientes = clientes.rename(columns={'Salario': 'Renda'})
```

```python
clientes.head(3)
```

```python
mapa_s = {'Masc': 'Masculino', 'Fem': 'Feminino'}

clientes['Sexo'] = clientes['Sexo'].map(mapa_s)
clientes.head(3)
```

```python
clientes['Media'] = clientes['Media'].fillna(0).astype(int)
```

```python
clientes['Media'].dtype
```

```python
notas = [
    clientes['Media'] > 80,
    (clientes['Media'] > 50) & (clientes['Media'] <= 80),
    clientes['Media'] <= 50
]
classi = ['Promotor', 'Neutro', 'Detrator']

clientes['NPS'] = np.select(notas, classi, default='unknown')

clientes.head(4)
```
